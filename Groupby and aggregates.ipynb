```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Dell E7450\Downloads\Python tutorial\Flavors.csv")
df
```

```python
group = df.groupby("Liked")
```

```python
for (Base Flavor), (Base Flavor)_df in group:
    print(Base Flavor)
    print(Base Flavor)_df
```

```python
for Liked, Liked_df in group:
    print(Liked)
    print(Liked_df)
```

```python
group.get_group('Yes')
```

```python
new_group = df.groupby('Base Flavor')
```

```python
new_group.get_group('Vanilla')
```

```python
 df.groupby('Base Flavor').mean(numeric_only = True)
```

```python
df.groupby('Base Flavor').describe( )
```

```python
df.groupby('Flavor Rating').agg({'Flavor Rating':['sum','count']})
```

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\Dell E7450\Downloads\Python tutorial\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"C:\Users\Dell E7450\Downloads\Python tutorial\LOTR 2.csv")
df2
```

```python
df1.merge(df2)
```

```python
df1.merge(df2, how = 'inner', on = 'FellowshipID')
```

```python
df1.merge(df2, how = 'outer', on = 'FellowshipID')
```

```python
df1.merge(df2, how = 'outer', on = 'FellowshipID',suffixes = ['_xb','_xy'])
```

```python
df1.merge(df2, how = 'outer', on = ['FellowshipID','FirstName'])
```

```python
df1.merge(df2, how = 'left', on = ['FellowshipID','FirstName'])
```

```python
df1.merge(df2, how = 'cross')
```

```python
df1.join(df2, how = 'outer', on = 'FellowshipID', lsuffix = '_Left', rsuffix = '_Right')
```

```python
df3= df1.set_index("FellowshipID").join (df2.set_index("FellowshipID"),lsuffix = '_Left', 
                                         rsuffix = '_Right')

df3
```

```python
pd.concat([df1,df2],axis =1)
```

```python
pd.concat([df1,df2],join = 'inner')
```
